```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mobile=pd.read_csv('https://github.com/sukhioo7/dataset/blob/main/Mobile%20Price%20Classification/train.csv?raw=true')
```

```python
mobile.head()
```

```python
mobile['price_range'].value_counts()
```

```python
mobile_corr=mobile.corr()
```

```python
plt.figure(figsize=(20,7))
sns.heatmap(mobile_corr,cmap='cool',annot=True)
```

```python
X = mobile.drop('price_range',axis=1)
```

```python
Y=mobile['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=4)
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```
